Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48

````python
ogrenciler = {
    "Ahmet":[72, 66, 48],
    "Ali":[82, 36, 88],
    "Elif":[56, 91, 84],
    "Nisa":[18, 24, 48],
    "Yagmur":[98, 45, 89]
}
ogrenci = input("Lutfen notlarini gormek istediginiz ogrencinin ismini giriniz")
ortalama = (ogrenciler[ogrenci][0]+ ogrenciler[ogrenci][1]+ ogrenciler[ogrenci][2])/3
print(
    ogrenci + " isimli ogrencinin 1 sinav notu "  +   str(ogrenciler[ogrenci][0])+
                                        " 2'ci sinav notu "  +  str(ogrenciler[ogrenci][1])+                                         
                                        " 3'cu sinav notu "  +  str(ogrenciler[ogrenci][2])+
                                        " ortalamasi "  + str(ortalama)
)
````
